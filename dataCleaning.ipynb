```python
import pandas as pd
import numpy
import os

bronx_df = pd.read_csv('./HourlyData/MTA_Subway_Hourly_Ridership_Bronx_All.csv', sep=',', low_memory=False)
brooklyn_df = pd.read_csv('./HourlyData/MTA_Subway_Hourly_Ridership_Brooklyn_All.csv', sep=',', low_memory=False)
manhattan_df = pd.read_csv('./HourlyData/MTA_Subway_Hourly_Ridership_Manhattan_All.csv', sep=',', low_memory=False)
queens_df = pd.read_csv('./HourlyData/MTA_Subway_Hourly_Ridership_Queens_All.csv', sep=',', low_memory=False)
staten_df = pd.read_csv('./HourlyData/MTA_Subway_Hourly_Ridership_StatenIsland_All.csv', sep=',', low_memory=False)
```

```python
bronx_df['transit_timestamp'] = pd.to_datetime(bronx_df['transit_timestamp'])
brooklyn_df['transit_timestamp'] = pd.to_datetime(brooklyn_df['transit_timestamp'])
manhattan_df['transit_timestamp'] = pd.to_datetime(manhattan_df['transit_timestamp'])
queens_df['transit_timestamp'] = pd.to_datetime(queens_df['transit_timestamp'])
staten_df['transit_timestamp'] = pd.to_datetime(staten_df['transit_timestamp'])

bronx_df['ridership'] = pd.to_numeric(bronx_df['ridership'], errors='coerce').fillna(0).astype(int)
bronx_df_day = bronx_df.set_index('transit_timestamp').groupby(['station_complex_id']).resample('D').ridership.sum()

brooklyn_df['ridership'] = pd.to_numeric(brooklyn_df['ridership'], errors='coerce').fillna(0).astype(int)
brooklyn_df_day = brooklyn_df.set_index('transit_timestamp').groupby(['station_complex_id']).resample('D').ridership.sum()

manhattan_df['ridership'] = pd.to_numeric(manhattan_df['ridership'], errors='coerce').fillna(0).astype(int)
manhattan_df_day = manhattan_df.set_index('transit_timestamp').groupby(['station_complex_id']).resample('D').ridership.sum()

queens_df['ridership'] = pd.to_numeric(queens_df['ridership'], errors='coerce').fillna(0).astype(int)
queens_df_day = queens_df.set_index('transit_timestamp').groupby(['station_complex_id']).resample('D').ridership.sum()

staten_df['ridership'] = pd.to_numeric(staten_df['ridership'], errors='coerce').fillna(0).astype(int)
staten_df_day = staten_df.set_index('transit_timestamp').groupby(['station_complex_id']).resample('D').ridership.sum()
```

```python

nyc_df = pd.concat([bronx_df_day, brooklyn_df_day, manhattan_df_day, queens_df_day, staten_df_day])
print(nyc_df.head())
nyc_df.to_csv('NYC_Daily_Ridership_2020-2024.csv')
```
