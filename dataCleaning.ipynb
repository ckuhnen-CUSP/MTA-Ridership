```python
import pandas as pd
import numpy
import os

bronx_df = pd.read_csv('./HourlyData/MTA_Subway_Hourly_Ridership_Bronx_All.csv', sep=',', low_memory=False)
```

```python
bronx_df['transit_timestamp'] = pd.to_datetime(bronx_df['transit_timestamp'])
bronx_df['ridership'] = pd.to_numeric(bronx_df['ridership'], errors='coerce').fillna(0).astype(int)
bronx_df_day = bronx_df.set_index('transit_timestamp').groupby(['station_complex_id']).resample('D').ridership.sum()
bronx_df.head()
```
